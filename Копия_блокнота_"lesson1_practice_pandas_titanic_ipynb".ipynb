Категоризируйте возраст пассажиров "Титаника" на три группы:

1. **Категория 1**: Молодые пассажиры (до 30 лет)
   - Если возраст меньше 30, функция возвращает 1.

2. **Категория 2**: Пассажиры среднего возраста (от 30 до 54 лет)
   - Если возраст от 30 до 54 включительно, функция возвращает 2.

3. **Категория 3**: Пожилые пассажиры (55 лет и старше)
   - Для всех остальных случаев (то есть 55 лет и старше) функция возвращает 3.

```python
def categorized_passengers_for_age(age):
  if pd.isna(age):  # Проверка на пропущенные значения
    return None
  if age < 30:
    return 1
  elif age >= 30 and age < 54:
    return 2
  elif age >= 55:
    return 3

df['age_category'] = df['Age'].apply(categorized_passengers_for_age)

# Вывод первых строк для проверки
print(df[['Age', 'age_category']].head())
```

**2. Выведите распределение переменной `Pclass` (социально-экономический статус) и это же распределение, только для мужчин / женщин по отдельности. Сколько было мужчин 2-го класса?**
- 104
- 108
- 112
- 125

```python
# Распределение Pclass для всех пассажиров
pclass_distribution = df['Pclass'].value_counts()
print("Распределение Pclass для всех пассажиров:")
print(pclass_distribution)

# Распределение Pclass для мужчин
male_pclass_distribution = df[df['Sex'] == 'male']['Pclass'].value_counts()
print("\nРаспределение Pclass для мужчин:")
print(male_pclass_distribution)

# Распределение Pclass для женщин
female_pclass_distribution = df[df['Sex'] == 'female']['Pclass'].value_counts()
print("\nРаспределение Pclass для женщин:")
print(female_pclass_distribution)

# Количество мужчин 2-го класса
men_second_class = male_pclass_distribution.get(2, 0)
print("\nКоличество мужчин 2-го класса:", men_second_class)
```
