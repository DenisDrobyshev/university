## **Задание №1. Базовая векторизация**


Напишите функцию, которая принимает два массива NumPy: вектор a размера (m,) и вектор b размера (m,). Функция должна вычислить косинусное расстояние между векторами (1 - (a·b)/(|a|·|b|)) без использования циклов Python.

```python
 import numpy as np

 def cosine_distance(a, b):
     cos = np.dot(a, b) / (np.linalg.norm(a) * np.linalg.norm(b))
     distance = 1 - cos
     return distance

print(cosine_distance(np.array([1, 2, 3]), np.array([4, 5, 6])))
```

## **Задание №2. Маскирование данных**


Создайте одномерный массив из 15 случайных целых чисел в диапазоне от -20 до 20. Используя маски, замените все положительные четные числа на 1, а все отрицательные числа умножьте на -0.5.

```python
import numpy as np

arr = np.random.randint(-20, 21, 15)
positive_mask = (arr > 0) & (arr % 2 == 0)
arr[positive_mask] = 1
negative_mask = (arr < 0)
arr[negative_mask] *= -0.5

print(arr)
```

## **Задание №3. Работа со структурированными массивами**


Создайте структурированный массив, содержащий информацию о 6 студентах с полями 'имя', 'возраст' и 'оценки' (массив из трех чисел). Вычислите средний балл каждого студента и добавьте это значение в новое поле 'средний_балл'.

```python
import numpy as np

student_data = np.zeros(6, dtype={'names': ('имя', 'возраст', 'оценки', 'средний_балл'),
                          'formats': ('U10', 'i4', '3i4', 'f8')})
student_data['имя'] = ['Deniska', 'Bob', 'Charlie', 'David', 'Eve', 'Frank']
student_data['возраст'] = [18, 19, 20, 18, 19, 20]
student_data['оценки'] = [(80, 85, 90), (70, 75, 80), (90, 95, 100), (60, 65, 70), (85, 90, 95), (75, 80, 85)]
student_data['средний_балл'] = np.mean(student_data['оценки'], axis=1)

print(student_data)
```

## **Задание №4. Преобразование формы массива**


Сгенерируйте одномерный массив из 30 последовательных чисел начиная с 5. Преобразуйте его в трехмерный массив размера 2×5×3. Найдите максимальное значение по каждому "срезу" вдоль второй оси и постройте из них новый массив.


```python
data = np.arange(5, 31)

resized_arr = np.resize(data, (2, 5, 3))
resized_arr = np.max(resized_arr, axis=1)

print(resized_arr)
```
